```python
#  Import required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#  Set visualization style
sns.set(style="whitegrid")

#  Load the dataset
df = pd.read_csv("survey_data_1000.csv")

#  Quick view of data
print("First 5 rows:")
print(df.head())
print("\nDataset Info:")
print(df.info())

#  Optional: Clean the data
df.drop_duplicates(inplace=True)
df.dropna(inplace=True)

#  Countplot - Response Type Distribution
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='ResponseType', palette='pastel')
plt.title('Response Count by Type')
plt.xlabel('Response Type')
plt.ylabel('Count')
plt.tight_layout()
plt.savefig("countplot_response_type.png")
plt.show()

#  Pie Chart - Satisfaction Level Distribution
satisfaction_counts = df['SatisfactionLevel'].value_counts().sort_index()
plt.figure(figsize=(6, 6))
plt.pie(
    satisfaction_counts,
    labels=satisfaction_counts.index,
    autopct='%1.1f%%',
    colors=sns.color_palette('pastel')
)
plt.title('Satisfaction Level Distribution')
plt.tight_layout()
plt.savefig("piechart_satisfaction_level.png")
plt.show()

#  Bar Plot - Average Satisfaction Level per Department
avg_satisfaction = df.groupby('Department')['SatisfactionLevel'].mean().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
sns.barplot(x=avg_satisfaction.values, y=avg_satisfaction.index, palette='viridis')
plt.title('Average Satisfaction Level by Department')
plt.xlabel('Average Rating')
plt.ylabel('Department')
plt.tight_layout()
plt.savefig("barplot_avg_satisfaction.png")
plt.show()
```
